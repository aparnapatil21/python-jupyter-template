### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11, 1)
y = x ** 2  # y values as the square of x

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

# Customize the plot
plt.title("Square of X Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values (X²)")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
